```python
import pandas as pd

left = pd.DataFrame({'key': ['k0', 'k1', 'k2'], 'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']}, index=['one', 'two', 'three'])
left
```

```python
right = pd.DataFrame({'key': ['k0', 'k1', 'k2'], 'C': ['c0', 'c1', 'c2'],
                     'D': ['d0', 'd1', 'd2']}, index=['one', 'two', 'three'])
right
```

```python
pd.merge(left, right)
```

```python
left2 = pd.DataFrame({'key': ['k0', 'k1', 'k2'], 'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})
left2
```

```python
right2 = pd.DataFrame({'key': ['k0', 'k1', 'k2'], 'A': ['a3', 'a4', 'a5'],
                     'B': ['b3', 'b4', 'b5']})
right2
```

```python
pd.merge(left2, right2, how='outer')
```

```python
pd.merge(left2, right2, how='inner')
```

```python
pd.merge(left2, right2, how='outer')
```

```python
pd.merge(left2, right2, how='left')
```

```python
pd.merge(left2, right2, how='right')
```

```python
pd.merge(left, right, on='key')
```

```python
left3 = pd.DataFrame({'key1': ['k0', 'k1', 'k2', 'k3'], 'key2': ['k0', 'k1', 'k2', 'k2'],
                      'A': ['a0', 'a1', 'a2', 'a3'], 'B': ['b0', 'b1', 'b2', 'b3']},
                    index=['one', 'two', 'three', 'four'])
left3
```

```python
right3 = pd.DataFrame({'key1': ['k0', 'k1', 'k1', 'k2'], 'key2': ['k0', 'k1', 'k1', 'k3'],
                      'C': ['c0', 'c1', 'c2', 'c3'], 'D': ['d0', 'd1', 'd2', 'd3']},
                    index=['one', 'two', 'three', 'four'])
right3
```

```python
pd.merge(left3, right3, on=['key1', 'key2'])
```

```python
pd.merge(left, right, left_on='key', right_on='key')
```

```python
pd.merge(left, right, left_on='key', right_on='C')
```

```python
pd.merge(left, right, left_on='key', right_on='C', how='outer')
```

```python
import numpy as np

pd.merge(left, right, left_on=np.array(['c0', 'c1', 'c2']), right_on='C')
```

```python
pd.merge(left, right, left_index=True, right_index=True)
```

```python
pd.merge(left, right, left_on='key', right_index=True, how='outer')
```

```python
pd.merge(left3, right3, on='key1')
```

```python
pd.merge(left3, right3, on='key1', suffixes=('_left', '_right'))
```

```python
pd.merge(left3, right3, on='key1', indicator=True)
```

```python
pd.merge(left3, right3, on=['key1', 'key2'], indicator='key_col')
```

```python
pd.merge(left3, right3, left_on='key1', right_index=True, 
         how='outer', indicator='key_col')
```

```python
pd.merge(left, right, on='key', validate='one_to_one')
```

```python
left4 = pd.DataFrame({'key': ['k0', 'k0', 'k0'], 'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']}, index=['one', 'two', 'three'])
left4
```

```python
right4 = pd.DataFrame({'key': ['k0', 'k1', 'k2'], 'C': ['c0', 'c1', 'c2'],
                     'D': ['d0', 'd1', 'd2']}, index=['one', 'two', 'three'])
right4
```

```python
pd.merge(left4, right4, on='key', validate='many_to_one')
```

```python
pd.merge(left4, right4, on='key', validate='many_to_many')
```
